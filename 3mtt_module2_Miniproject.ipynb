```python
#splitting the data into two distinct groups
#cat_df = categorical columns, categorical refers to object data types which are the same thing as strings
#num_df = numerical columns

cat_df = df.select_dtypes(include=['object']) #returns dataframe that contain the object data type
num_df = df.select_dtypes(include=['number']) #returns columns that contain the numerical data type

cat_df.head()

```

```python
#creating functions

#discrete = categorical
#univariate = one

def discrete_uni_analysis(data):
    for col_name in data:
        print('\n','*'*5, col_name, '*'*5) #for formatting
        print(data[col_name].agg(['count', 'nunique', 'unique']))
        print('Value_counts: \n', data[col_name].value_counts())

#count the number of the rows of data
#nunique = the number of unique items in the column
#unique = find out unique items
#value counts = for each unique item in our column, how many times do they appear

def num_uni_analysis(data):
    for col_name in data:
        print('\n','*'*5, col_name, '*'*5)
        print(data[col_name].agg(['count', 'min', 'max', 'mean','skew', 'median', 'std']))
```

```python
#using the functions
discrete_uni_analysis(cat_df)
```

```python
num_uni_analysis(num_df)
```

#**Step 5: Visualization, Identify Trends and Insights based on the Formulate Research Questions(From Step 2) for the Dataset**

###**Revenue - focused questions**
1. What are the top-selling products by revenue and quantity: Ths will highlight key revenue drivers and potential areas for growth.
2. Which customer segments generate the most revenue: This will focus on marketing efforts on high- value customers.
3. Are there any seasonal trends in sales for specific product categories over time: This will optimize inventory and marketing strategies by examining patterns.

```python
# Revenue- focused question
#Question 1: The top - selling product by revenue and quantity

# Top-selling products by Revenue
top_products_revenue = df.groupby('Product')['Total'].sum().sort_values(ascending=False).head(10)

# Top-selling products by Quantity
top_products_quantity = df.groupby('Product')['Quantity'].sum().sort_values(ascending=False).head(10)

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(10, 4))  # 1 row, 2 columns

# Plot 1: Top products by Revenue
sns.barplot(x=top_products_revenue.index, y=top_products_revenue.values, ax=axes[0])
axes[0].set_title('Top Selling Products by Revenue')
axes[0].set_xlabel('Product')
axes[0].set_ylabel('Total Revenue')

# Plot 2: Top products by Quantity
sns.barplot(x=top_products_quantity.index, y=top_products_quantity.values, ax=axes[1])
axes[1].set_title('Top Selling Products by Quantity')
axes[1].set_xlabel('Product')
axes[1].set_ylabel('Total Quantity Sold')

plt.tight_layout()  # Adjust spacing between subplots
plt.show()

#IDENTIFY PATTERN
##To identify the top-selling product in revenue and total

#INSIGHT: Based on the visualization,
##Laptop is the most purchased product based on quantity and contributes greatly to sales revenue growth. This indicates a high demand for laptops among customers
##This is followed by phone and tablet
```

```python
#Question 2:Which customer segments generate the most revenue

#First we group and aggregate
# Group by CustomerID and sum Total revenue for each customer
customer_revenue = df.groupby('CustomerID')['Total'].sum().sort_values(ascending=False)

# Select the top 5 customers
top_customers = customer_revenue.head(10)

# Create a bar plot to visualize the top 10 customers
plt.figure(figsize=(10, 5))
sns.barplot(x=top_customers.index, y=top_customers.values)
plt.title('Top 10 Customers by Revenue')
plt.xlabel('Customer ID')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()

#IDENTFY PATTERN
##From my visualization, the high- value customers with the highest purchase revenue is C003, followed by C001, C016. Based on the chart, they are slightly similar to each other

#INSIGHT
##Customer C003, C001, and C016 are the most valuable customers in terms of revenue generation.
```

```python
#Question 3: Are there any seasonal trends in sales for specific product categories over time and month sales revenue trend
import calendar

# Convert 'Date' to datetime and extract month
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month  # Extract month as number

# Create a custom month order dictionary
month_order = {i: calendar.month_name[i] for i in range(1, 13)}

# Group by month and product category, sum total revenue
monthly_product_revenue = df.groupby(['Month', 'Product'])['Total'].sum().reset_index()

# Reindex to ensure months are in ascending order
monthly_product_revenue['Month_Name'] = monthly_product_revenue['Month'].map(month_order)
monthly_product_revenue = monthly_product_revenue.sort_values('Month')

# Monthly sales revenue trend
monthly_revenue = df.groupby('Month')['Total'].sum().reindex(month_order).reset_index()
monthly_revenue['Month'] = monthly_revenue['Month'].map(month_order)

# Create subplots for monthly sales trends and overall revenue trend
fig, axes = plt.subplots(2, 1, figsize=(12, 10))  # 2 rows, 1 column

# Plot 1: Monthly Sales Trends for Product Categories
sns.lineplot(data=monthly_product_revenue, x='Month_Name', y='Total', hue='Product', ax=axes[0], marker="o")
axes[0].set_title('Monthly Sales Trends for Product Categories')
axes[0].set_xlabel('Month')
axes[0].set_ylabel('Total Sales')
axes[0].set_xticks(monthly_product_revenue['Month_Name'].unique())  # Set x-axis ticks to represent months
axes[0].legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # Place legend outside the plot

# Plot 2: Overall Monthly Sales Revenue Trend
sns.lineplot(data=monthly_revenue, x='Month', y='Total', ax=axes[1], marker="o")
axes[1].set_title('Monthly Sales Revenue Trend')
axes[1].set_xlabel('Month')
axes[1].set_ylabel('Total Revenue')
axes[1].set_xticks(monthly_revenue['Month'].unique())  # Set x-axis ticks to represent months

plt.tight_layout()  # Adjust spacing between subplots
plt.show()

#IDENTIFY PATTERN
#Seaonal Trend
##To identify any recurring patterns or peaks during specific months or seasons
##To identify specific seasonal patterns for individual product

#Month Sales Revenue Trend
##To identify the months with the lowest sales revenue
## identify the months with the highest sales revenue.

#INSIGHT:Based on the visualizaton
#Seasonal Trend:The overall recurrng patterns is mostly across June to December
##Laptop has its peak sales in the month of December, June and April
##Phone has its peak sales in August, May and February
##Tablet has its peak sales in July and October

#Monthly Sales Revenue Trend
##The peak is mostly from February which later decline and then peak again from August to December
```

###**Customer satisfaction-focused questions**
1. What are the purchase frequencies and recency for different customer segments using Date and orderId:This can help understand customer engagement and loyalty.
2. Can we identify any customer segments that are more likely to purchase specific product categories: This can inform targeted marketing efforts.


```python
#Question 1: What are the purchase frequencies and recency for different customer segments using Date and orde
# Convert 'Date' to datetime:
df['Date'] = pd.to_datetime(df['Date'])

# 2. Group data by CustomerID and Date to get the frequency of purchases:
frequency_df = df.groupby(['CustomerID', 'Date'])['OrderID'].count().reset_index()
frequency_df.rename(columns={'OrderID': 'Frequency'}, inplace=True)  # Rename 'OrderID' to 'Frequency'

# 3. Create the line plot using Seaborn:
plt.figure(figsize=(12, 6))  # Adjust figure size as needed
sns.lineplot(data=frequency_df, x='Date', y='Frequency', hue='CustomerID', marker='o')
plt.title('Customer Purchase Frequency Over Time')
plt.xlabel('Date')
plt.ylabel('Purchase Frequency')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.legend(title='CustomerID', bbox_to_anchor=(1.05, 1), loc='upper left') # Place legend outside the plot
plt.tight_layout()
plt.show()

#IDENTIFY PATTERN
##Customers with higher frequency tend to have lower recency

#INSIGHT:Based on this visualization
##Customer C002 has the highest purchase frequency and low recency
```

```python
#Question 2: #Question 2: Can we identify any customer segments that are more likely to purchase specific product categories

# Group and aggregate the data by customerId , product and Quantty
customer_product_quantity = df.groupby(["CustomerID", "Product"])["Quantity"].sum().reset_index()

# Pivot the data to create a matrix for the heatmap
customer_product_matrix = customer_product_quantity.pivot(index='CustomerID', columns='Product', values='Quantity').fillna(0)

# Create the heatmap for visualization
plt.figure(figsize=(10, 6))
sns.heatmap(customer_product_matrix, annot=True, cmap="YlGnBu", fmt=".2f", cbar_kws={'label': 'Total Quantity Purchased'})
plt.title('Customer and Their Preferred Product Categories')
plt.xlabel('Product Category')
plt.ylabel('Customer ID')
plt.tight_layout()
plt.show()

#IDENTFY PATTERNS:
##Darker colors represent higher quantities purchased indicating a stronger preference for that product category by the customer.
##Lighter colors represent lower quantities or no purchases, indicating less preference or no interest in that category.

#INSIGHTS:Based on this visualization
## Customer C003 has darker colours across laptop, phone and tablet, indicating a strong preference for electronic product
##Customers C001, C006, C003, C004, C009,C016 and C017 has a darker colour for laptop, indicating strong interest for laptop purchase
##Customer C005 has a darker colour for phone thus indicating its interest in phone purchase while C020 has a slightly darker colour thus it slightly interest
##Customer C003 has a darker colour fot tablet, this shows interest in tablets purchase
```
